```
loclist = []
for index, row in locs.iterrows():
  loclist.append({'lat': row['Location'][0]['geometry']['location']['lat'], 
                  'lng': row['Location'][0]['geometry']['location']['lng'],
                  'IK-SO': row['IK-SO']})
print(len(loclist))
```

```
from geoindex import GeoGridIndex, GeoPoint
def getClinicsDis(address, dist):
  location = gmaps.geocode(address)
  index = GeoGridIndex()
  for IK in loclist:
      index.add_point(GeoPoint(IK['lat'], IK['lng'], ref=IK["IK-SO"]))

  center_point = GeoPoint(location[0]['geometry']['location']['lat'], location[0]['geometry']['location']['lng'])
  for distance, point in index.get_nearest_points(center_point, dist, 'km'):
    print("We found a clinic in {0}km at {1}".format(point, distance))
```
